## Advanced LaneLine Finding

### Camera Calibration
The camera calibration aims to calculate the Matrix and Dist values in order to be used in Undistorison. The steps are the following:
1. List all images names
2. Per each image find the chessboard corners and append those corners
3. Calculate the Matrix and Dist Values

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
%matplotlib inline

## Extracting all images names
images = glob.glob('../media/calibration_images/calibration*.jpg')

## Creating Matrix for objpoints and imgpoints
objpoints = [] # 3D Points
imgpoints = [] # 2D Points

## Number of Corners (x, y) 
nx = 9
ny = 5

## Create the objects points
objp = np.zeros((nx*ny,3), np.float32)
objp[:,:2] = np.mgrid[0:nx,0:ny].T.reshape(-1,2) # x,y coordinates

## Corners calculation per image
for fname in images:
    ## Image Reading
    img = cv2.imread(fname)
    ## BGR2GRAY Conversion
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ## Corners Finding
    ret , corners = cv2.findChessboardCorners(gray, (nx, ny), None)
    if ret == True:
        ## Image and Object Points appending
        imgpoints.append(corners)
        objpoints.append(objp)

## Taking Image Size
img_size = (img.shape[1], img.shape[0])

## Calculate the Calibration Matrix 
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, img_size,None,None)
```

### Camera undistorsion
We have already calculated Matrix and Dist value, so here we will define the Undistorsion program and a shot example of how does it works

```python
## Undistorsion Function
def Undistorsion(img):
    return cv2.undistort(img, mtx, dist, None, mtx)

## Image Reading
img = cv2.imread('../media/calibration_images/calibration1.jpg')    
## Image Undistorsion
undistorted_img = Undistorsion(img)
## Distorted and Undistorted Image
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(img)
ax1.set_title('Original Image', fontsize=50)
ax2.imshow(undistorted_img)
ax2.set_title('Undistorted Image', fontsize=50)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

### Image Thresholding for Line Detection


```python
## Sobel Filter
def SobelFilter(img, orient='x',sobel_kernel= 3, thresh=(0, 255)):
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    if orient == 'x':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0))
    elif orient == 'y':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 0, 1))
    elif orient == 'xy':
        sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
        sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
        abs_sobel = np.sqrt(sobelx**2 + sobely**2)
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    sbinary = np.zeros_like(scaled_sobel)
    sbinary[(scaled_sobel >= thresh[0]) & (scaled_sobel <= thresh[1])] = 1
    binary_output = sbinary
    return binary_output

## Thresh Filter
def ThreshFilterHS(img, thresh_S = (100, 255), thresh_H = (100, 255)):
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    H = hls[:,:,0]
    L = hls[:,:,1]
    S = hls[:,:,2]

    binary_output = np.zeros_like(S)
    binary_output[(S > thresh_S[0]) & (S <= thresh_S[1]) & (H < thresh_H[1]) &  (H > thresh_H[0])] = 1
    return binary_output

## Full Filtering
def ImageFilter(img):
    kernel_size = 3
    mag_threshMag = (30,160)
    thresh_S = (100, 200)
    thresh_H = (100, 200)
    
    sobelMag = SobelFilter(img,'xy',kernel_size, mag_threshMag)
    HS_bin = ThreshFilterHS(img, thresh_S, thresh_H)
    combined = np.zeros_like(sobelMag)
    combined[((sobelMag == 1) ) | (HS_bin == 1)] = 1
    return combined


img = cv2.imread('../media/test_images/test1.jpg')    
img = Undistorsion(img)

combined = ImageFilter(img)
plt.imshow(img)
plt.show()
plt.imshow(combined, cmap='gray')
plt.show()
```

### Region Selection

```python
def RegionSelection(image):
    ## First me make a copy of the image
    region_select = np.copy(image)
    xsize = image.shape[1]
    ysize = image.shape[0]
    print (xsize, ysize)
    ## Second we define corners of the triangle containing the road
    left_bottom = [1*xsize/20,ysize]
    right_bottom = [19*xsize/20, ysize]
    apex_right = [11*xsize/20, 11 * ysize/20]
    apex_left = [10*xsize/20, 11 * ysize/20]
    ## Lines Creation
    fit_left = np.polyfit((left_bottom[0],apex_left[0]),(left_bottom[1],apex_left[1]),1)
    fit_right = np.polyfit((right_bottom[0],apex_right[0]),(right_bottom[1],apex_right[1]),1)
    fit_bottom = np.polyfit((left_bottom[0],right_bottom[0]),(left_bottom[1],right_bottom[1]),1)
    fit_top = np.polyfit((apex_left[0],apex_right[0]),(apex_left[1],apex_right[1]),1)
    XX, YY = np.meshgrid(np.arange(0, xsize), np.arange(0, ysize))
    ## Region Thresholding
    region_thresholds =  (YY < XX*fit_left[0] + fit_left[1]) | (YY < XX*fit_right[0] + fit_right[1]) | (YY < XX*fit_top[0] + fit_top[1])
    region_select[region_thresholds] = 0
    return region_select

reg_selected = RegionSelection(combined)
plt.imshow(reg_selected, cmap='gray')
plt.show()
```

### Image Warping

```python
## Show the points you want to rotate
plt.imshow(reg_selected, cmap='gray')
plt.plot(230, 700, ".")
plt.plot(590, 450, ".")
plt.plot(690, 450, ".")
plt.plot(1070, 700, ".")
plt.show()

def Warp(img):
    img_size = (img.shape[1], img.shape[0])
    src = np.float32(
        [[230,700],
         [590,450],
         [690,450],
         [1100,700] 
        ])
    dst = np.float32(
        [[230,700],
         [230,50],
         [1100,50],
         [1100,700] 
        ])
    M =  cv2.getPerspectiveTransform(src, dst)
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    return warped

def Unwarp(img):
    img_size = (img.shape[1], img.shape[0])
    dst = np.float32(
        [[230,700],
         [590,450],
         [690,450],
         [1100,700] 
        ])
    src = np.float32(
        [[230,700],
         [230,50],
         [1100,50],
         [1100,700] 
        ])
    M =  cv2.getPerspectiveTransform(src, dst)
    unwarped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    return unwarped



warped_im =  Warp(combined)
plt.imshow(warped_im, cmap="gray")
```

### Road Calculation
For that purpose we use an histogram

```python
import numpy as np

def RoadCalculation(warped_im):
    # Calculate the sum of ones per each colum
    histogram = np.sum(warped_im[warped_im.shape[0]//2:,:], axis=0)
    # Calculate the colums midpoint
    midpoint = np.int(histogram.shape[0]/2)
    # Calculate the index where you find the max value of the histogram
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint
    # Place together 3 layers
    out_img = np.dstack((warped_im, warped_im, warped_im))*255
    # Choose the number of sliding windows
    nwindows = 20
    # Set height of windows
    window_height = np.int(warped_im.shape[0]/nwindows)
    # Get the non Zero pixels positions
    nonzero = warped_im.nonzero()
    # Extract the y index of  non-Zero Positions
    nonzeroy = np.array(nonzero[0])
    # Extract the x index of non-Zero Positions
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated for each window
    leftx_current = leftx_base
    rightx_current = rightx_base
    # Set the width of the windows +/- margin
    margin = 50
    # Set minimum number of pixels found to recenter window
    minpix = 30
    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []
    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = warped_im.shape[0] - (window+1)*window_height
        win_y_high = warped_im.shape[0] - window*window_height
        win_xleft_low = leftx_current - margin
        win_xleft_high = leftx_current + margin
        win_xright_low = rightx_current - margin
        win_xright_high = rightx_current + margin
        # Draw the windows on the visualization image
        # Identify the nonzero pixels in x and y within the window
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & (nonzerox >= win_xleft_low) & (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & (nonzerox >= win_xright_low) & (nonzerox < win_xright_high)).nonzero()[0]
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        # If you found > minpix pixels, recenter next window on their mean position
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))

    # Concatenate the arrays of indices
    left_lane_inds = np.concatenate(left_lane_inds)
    right_lane_inds = np.concatenate(right_lane_inds)

    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds] 

    # Fit a second order polynomial to each
    left_fit = np.polyfit(lefty, leftx, 2)
    right_fit = np.polyfit(righty, rightx, 2)

    # Generate x and y values for plotting
    ploty = np.linspace(0, warped_im.shape[0]-1, warped_im.shape[0] )

    # Calculate the left an right fits
    left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]

    left_line = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    right_line = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    road = np.hstack((left_line, right_line))
    
    superposed_image = np.zeros_like(out_img)
    
    ##Draw the filling
    cv2.fillPoly(superposed_image, np.int_([road]), (0,255, 0))
    y_eval = np.max(ploty)

    ## Left and Right Curvature Calculation
    left_curverad = ((1 + (2*left_fit[0]*y_eval + left_fit[1])**2)**1.5) / np.absolute(2*left_fit[0])
    right_curverad = ((1 + (2*right_fit[0]*y_eval + right_fit[1])**2)**1.5) / np.absolute(2*right_fit[0])
    
    return superposed_image

## Warped Superposed Image
superposed_image = RoadCalculation(warped_im)
plt.imshow(superposed_image)
plt.show()
## Unwarped Superposed Image
unwarped_superposed = Unwarp(superposed_image)
plt.imshow(unwarped_superposed)
plt.show()
## Weighted Superposed + Original
result = cv2.addWeighted(img, 1, unwarped_superposed, 0.3, 0)
plt.imshow(result)
plt.show()
```
